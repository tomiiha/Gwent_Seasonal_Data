```python
# Packages
import os
import pandas as pd
import pathlib
```

```python
def data_finder():
    path = 'season_data/'
    all_files = [x for x in os.listdir(path) if x.endswith('.csv')]
    main_df = pd.concat((pd.read_csv(path+x) for x in all_files),ignore_index=True)
    return main_df
```
